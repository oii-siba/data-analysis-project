```python
import pandas as pd

df = pd.read_csv('job_salary_rich_dataset_updated.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df_numeric = df.select_dtypes(include=['number'])

Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

mask = ~((df_numeric < (Q1 - 1.5 * IQR)) |
         (df_numeric > (Q3 + 1.5 * IQR))).any(axis=1)

df_clean = df[mask]

print("Before:", df.shape)
print("After:", df_clean.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(12,5))

# Before
plt.subplot(1,2,1)
sns.histplot(df['salary'], kde=True)
plt.title("Before Outlier Removal")

# After
plt.subplot(1,2,2)
sns.histplot(df_clean['salary'], kde=True)
plt.title("After Outlier Removal")

plt.show()
```

```python
sns.boxplot(data=df_numeric)
plt.title("Box Plot (Outlier Detection)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df_numeric.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df_clean['salary'], kde=True)
plt.title("Salary Distribution")
plt.show()
```

```python
plt.plot(df_clean['experience_years'], df_clean['salary'])
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Line Plot")
plt.show()
```

```python
df_clean['education_level'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Education Level Distribution")
plt.show()
```

```python
plt.hist(df_clean['salary'])
plt.title("Histogram : Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df_clean['experience_years'], df_clean['salary'])
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Scatter Plot")
plt.show()
```

```python
df_clean['job_title'].value_counts().plot.bar()
plt.title("Bar Graph")
plt.xticks(rotation=45)
plt.show()
```
